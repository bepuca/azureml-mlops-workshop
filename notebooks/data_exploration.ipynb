```python
from pathlib import Path


import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
dataset_path = Path("../data/mnist_sample")
threes_path = dataset_path / "3"
sevens_path = dataset_path / "7"
```

```python
threes_filenames = list(threes_path.glob("*"))
sevens_filenames = list(sevens_path.glob("*"))

duplicated_filenames = set(threes_filenames).intersection(set(sevens_filenames))
print(f"num filenames duplicated in both classes: {len(duplicated_filenames)}")

n_threes = len(threes_filenames)
n_sevens = len(sevens_filenames)
n_tot = n_threes + n_sevens

print(f"num 3s: {n_threes} ({100*n_threes/n_tot:.0f}%)")
print(f"num 7s: {n_sevens} ({100*n_sevens/n_tot:.0f}%)")
print(f"total num: {n_tot}")
```

```python
n_samples = 5
threes_sample = np.random.choice(threes_filenames, n_samples)
sevens_sample = np.random.choice(sevens_filenames, n_samples)

fig, axes = plt.subplots(nrows=2, ncols=n_samples, figsize=(10, 4))
for i, samples in enumerate([threes_sample, sevens_sample]):
    for j in range(n_samples):
        filename = samples[j]
        img = Image.open(filename)
        axes[i,j].imshow(img, cmap="gray")
        axes[i,j].set_title(filename.name)
        axes[i,j].axis("off")
```
